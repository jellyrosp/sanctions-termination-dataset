## Checking Frequencies Before Inferential Testing

### Cochran golden standard
- No cell should have an expected frequency < 1
- No more than 20% of cells should have an expected frequency < 5

```python
# expiry_x_duration
check_expected_frequencies(desc.get_expiry_x_duration(con))
expiry_x_duration = remove_rel_freq(desc.get_expiry_x_duration(con))

print("\n")
print("=" * 60)
print("NORMALIZED CONTINGENCY TABLE")
print("=" * 60)
print(expiry_x_duration)
print("=" * 60)
```
